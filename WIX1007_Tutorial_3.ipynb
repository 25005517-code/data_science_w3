1. Create a numpy array of size 2 x 2

```python
import numpy as np
```

```python
n1 = np.array([[1,2],[3,4]])
print(n1)
```

2. Create a random numpy array of size 3 x 4

```python
n2 = np.random.random((3,4))
print(n2)
```

3. Create a numpy array of ones and a numpy array of zeroes

```python
n3a = np.ones((2,2))
print(n3a)
```

```python
n3b = np.zeros((2,2))
print(n3b)
```

4. Create a random numpy array of size 6x3 and transpose it

```python
n4a = np.random.random((6,3))
print('A random numpy array: \n')
print(n4a)
```

```python
n4b = n4a.T
print('\n\nTransposed numpy array: \n')
print(n4b)
```

5. Create two random numpy array of size n x p and p x m and then compute the dot product

```python
n=3
p=5
m=4

n5a = np.random.random((n,p))
n5b = np.random.random((p,m))

n5c = np.dot(n5a, n5b)

print('First numpy array: \n\n', n5a)
print('\nSecond numpy array: \n\n', n5b)
print('\nThe dot product: \n\n', n5c)
```
